Кодирование методом One-Hot Encoding

```python
def one_hot_encode_columns(df, columns):
    encoder = OneHotEncoder(sparse_output=False, drop='first')
    
    encoded_data = encoder.fit_transform(df[columns])
    
    encoded_df = pd.DataFrame( encoded_data, columns=encoder.get_feature_names_out(columns), index=df.index )
    
    df_encoded = df.drop(columns=columns)
    df_encoded = pd.concat([df_encoded, encoded_df], axis=1)
    
    return df_encoded, encoder

df_onehot, onehot_encoder = one_hot_encode_columns(df, categorical_features)

df_onehot.head()
```

5. Создать дополнительные признаки (минимум 2 новых).

```python
def create_features(df):
    df = df.copy()
    
    if "vote_average" in df.columns and "vote_count" in df.columns:
        df["vote_ratio"] = df["vote_average"] / (df["vote_count"] + 1)
    
    if "popularity" in df.columns and "runtime" in df.columns:
        df["popularity_per_min"] = df["popularity"] / (df["runtime"] + 1)
    
    return df

df = create_features(df)
df[["vote_ratio", "popularity_per_min"]].head()
```

7. Подготовить финальный датасет для модели.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

def prepare_final_dataset_label(df):
    df = df.copy()
    
    irrelevant_columns = ["Unnamed: 0", "id", "tagline", "overview", "homepage"]
    df = df.drop(columns=irrelevant_columns, errors="ignore")
    
    if "vote_average" in df.columns and "vote_count" in df.columns:
        df["vote_ratio"] = df["vote_average"] / (df["vote_count"] + 1)
    if "popularity" in df.columns and "runtime" in df.columns:
        df["popularity_per_min"] = df["popularity"] / (df["runtime"] + 1)

    categorical_features = ["original_language", "genre"]
    label_encoders = {}
    
    for col in categorical_features:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col].astype(str))
        label_encoders[col] = le
    
    numerical_features = df.select_dtypes(include=["int64", "float64"]).columns.tolist()
    scaler = StandardScaler()
    df[numerical_features] = scaler.fit_transform(df[numerical_features])
    
    return df, label_encoders, scaler

final_df_label, label_encoders, final_scaler = prepare_final_dataset_label(df)

print("Финальный датасет с Label Encoding готов к использованию!")
print("Форма:", final_df_label.shape)
final_df_label.head()
```
